```python
import sys
import os
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd
from bs4 import BeautifulSoup
import re 
```

```python
def cleanReply(readData):    # 리뷰 정리
    result = re.search('<br.*', readData)
    result2 = result.group()
    result3 = re.sub('<br/>','', result2)
    return(result3)
```

```python
def cleanStar(readData):   # 별점 정리
    result = re.search('>.{2}',readData)
    result2 = result.group()
    result3 = re.sub('>','', result2)
    result4 = re.sub('<','', result3)
    return(result4)
```

```python
def save_reply(code):      
    #해당지점에 medicine1이라는 텍스트 파일을 열고 새로운 것을 이어적는다.
    f = open('./data/movie3.txt','a',encoding='utf-8') 
    f.write(code+"|")
    f.close()
```

```python
def save_star(code):      
    #해당지점에 medicine1이라는 텍스트 파일을 열고 새로운 것을 이어적는다.
    f = open('./data/movie3.txt','a',encoding='utf-8') 
    f.write(code+"\n")
    f.close()
```

```python
   
#부정적 리뷰만 추출
def copy():
    for i in range(1,11):
        time.sleep(1)
        
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        te = soup.select('#old_content > table > tbody > tr:nth-child({0}) > td.title'.format(i))
        ste = str(te)

        st = soup.select('#old_content > table > tbody > tr:nth-child({0}) > td.point'.format(i))
        sst = str(st)
        
        a = cleanReply(ste)
        b = cleanStar(sst)
        c = ""
        
        if int(b) <= 3:
            c = "0"
            save_reply(a)
            save_star(c)
        else:
            pass

            
        lines = open('./data/movie3.txt', 'r',encoding='UTF8').readlines()
        lines_set = set(lines)
        
        # 중복없앰
        out = open('./data/movie4.txt', 'w',encoding='UTF8')
        for line in lines_set:
            out.write(line)
            
        else:
            pass
```

```python
#다음페이지
def next_page():
        try:
            WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.CSS_SELECTOR,'#old_content > div.paging > div > a.pg_next'))).click()
            time.sleep(1)
        except Exception as e:
            pass
```

```python
def movie_cr(num):
    global driver
    driver = webdriver.Chrome()
    driver.get('https://movie.naver.com/movie/point/af/list.nhn?target=before')
    i = 0
    while i < num:
        copy()
        next_page()
        i = i+1
```
