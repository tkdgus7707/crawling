```python
def resampling(data):     # resampling 함수
    count_class_1, count_class_0 = data.score.value_counts()

    # Divide by class
    df_class_0 = data[data['score'] == 0]
    df_class_1 = data[data['score'] == 1]
    
    df_class_1_under = df_class_1.sample(count_class_0, replace=True)
    df_test_under = pd.concat([df_class_0, df_class_1_under], axis=0)
    
    return(df_test_under)
```

```python
data1 = resampling(data)
```
